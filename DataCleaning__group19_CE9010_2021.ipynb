# Cleaning data
<hr> Objective is to remove NaN and unnesccary columns</hr> 

### Import Libraries

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

### Read csv file

```python
df = pd.read_csv('./ScrapedData_group19_CE9010_2021.csv')
df
```

### Converting cells that contain '-' in csv file to NaN for row removal if a row consist of Nan

```python
df.replace('-', np.nan, inplace=True)
df.fillna(method='None')
df.info()
```

```python
df
```

### Seeing how many NaN in Price

```python
df.isna().sum()
```

### Drop NaN

```python
df = df.dropna()
df
```

### Converting cells that has 'N.A' to NaN

```python
df.replace('N.A', np.nan, inplace=True)
unnecessary_columns = ['Img_url','Eng Cap','Status']
df = df.drop(unnecessary_columns, axis=1)
df
```

```python
df.info()
```

```python
df.apply(pd.to_numeric, errors='ignore')

{column: len(df[column].unique())for column in df.columns if df.dtypes[column] == 'object'}

```

```python
df.to_csv('CleanedData_group19_CE9010_2021.csv')
```
